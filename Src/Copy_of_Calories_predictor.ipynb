```python
import pandas as pd
import numpy as np


data = {
    "Weight_kg": [60, 65, 70, 75, 80, 68, 72, 85, 90, 78],
    "Duration_min": [30, 45, 60, 40, 50, 35, 55, 60, 45, 50],
    "HeartRate": [110, 120, 130, 115, 125, 118, 128, 135, 122, 124],
    "Steps": [3500, 5000, 7000, 4500, 6000, 4200, 6500, 7200, 5200, 5800],
    "Calories": [150, 230, 320, 210, 290, 190, 305, 350, 260, 275]
}

df = pd.DataFrame(data)

print(df)

x=df.drop("Calories",axis=1)
y=df["Calories"]


from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)


x_train_scaled=x_train/np.max(x_train,axis=0)
x_test_scaled=x_test/np.max(x_train,axis=0)

w=np.random.randn(x_train_scaled.shape[1])
b=np.random.randn()


learning_rate=0.01
iterations=1000


for i in range(iterations):
  y_pred=x_train_scaled@w+b
  error=y_pred-y_train

  grad_w=x_train_scaled.T@error
  grad_b=np.mean(error)

  w=w-learning_rate*grad_w
  b=b-learning_rate*grad_b

  pred_test=x_test_scaled@w+b


  from sklearn.metrics import mean_squared_error
  from sklearn.metrics import mean_absolute_error
  from sklearn.metrics import r2_score


  mse=mean_squared_error(y_test,pred_test)
  mae=mean_absolute_error(y_test,pred_test)
  r2=r2_score(y_test,pred_test)

  print("MSE: ",mse)
  print("MAE: ",mae)
  print("R2: ",r2)


weights=int(input("Enter your weight:"))
duration=int(input("Enter the duration:"))
heart_rate=int(input("Enter your heartrate:"))
steps=int(input("Enter your steps:"))


new_person=np.array([[weights, duration,heart_rate,steps]])
new_person_scaled=new_person/np.max(x_train,axis=0).values

prediction=new_person_scaled@w+b
print("Predicted calories: ",prediction)
```
